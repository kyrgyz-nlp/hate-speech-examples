```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification",
                      model="joeddav/xlm-roberta-large-xnli")
```

Take hate speech examples in Kyrgyz, get scores given by the model and evaluate mismatch

```python
import csv

hate_speech_count = 0
accumulator = []
candidate_labels = ["hate speech", "neutral"]

def get_scores(model_prediction_dict):
    first_label, second_label = model_prediction_dict['labels']
    first_score, second_score = model_prediction_dict['scores']
    if first_label == 'neutral':
        # hate speech score, neutral score
        return [second_score, first_score]
    return [first_score, second_score]

!wget 'https://raw.githubusercontent.com/kyrgyz-nlp/hate-speech-examples/main/with_kyrgyz_translation.tsv'

with open('with_kyrgyz_translation.tsv') as f:
    reader = csv.reader(f, delimiter='\t')
    header = next(reader)
    for comment, hate_speech_score, translation in reader:
      translation_classification_result = classifier(translation, candidate_labels)
      hate_speech_score, neutral_score = get_scores(translation_classification_result)
      is_hate_speech = 1 if hate_speech_score > 0.5 else 0
      if is_hate_speech:
        hate_speech_count += 1
      accumulator.append([translation, is_hate_speech])
      print(translation, '\t', is_hate_speech)
print('hate_speech_count: ', hate_speech_count, 'all examples:', len(accumulator))
```

Take hate speech examples in Kyrgyz, convert them into sentences with typos (as they usually appear in social media) and get their hate speech scores given by the model and evaluate mismatch

```python
hate_speech_count = 0
accumulator = []
candidate_labels = ["hate speech", "neutral"]

!wget 'https://raw.githubusercontent.com/kyrgyz-nlp/hate-speech-examples/main/with_kyrgyz_translation_and_typo_version.tsv'

with open('with_kyrgyz_translation_and_typo_version.tsv') as f:
    reader = csv.reader(f, delimiter='\t')
    header = next(reader)
    for comment, hate_speech_score, translation, translation_with_typos in reader:
      translation_with_typos_classification_result = classifier(translation_with_typos, candidate_labels)
      hate_speech_score, neutral_score = get_scores(translation_with_typos_classification_result)
      is_hate_speech = 1 if hate_speech_score > 0.5 else 0
      if is_hate_speech:
        hate_speech_count += 1
      accumulator.append([translation, is_hate_speech])
      print(translation_with_typos_classification_result['sequence'], '\t', is_hate_speech)
print('hate_speech_count: ', hate_speech_count, 'all examples:', len(accumulator))
```
